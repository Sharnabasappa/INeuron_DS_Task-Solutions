```python
# 8  Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset and find out how many healthy person we have in our dataset
```

```python
# assumption 1800 calarie burn to be healty person
def caloriecheck(a):
    if a>= 1800:
        return 1
    else:
        return 0
```

```python
df.head(2)
```

```python
df['HealthyDay']= df['Calories'].apply(caloriecheck)  # new column'healty data created
```

```python
df[['HealthyDay', 'Calories']]
```

```python
health_df =  df.groupby(['Id'])[['HealthyDay','Calories' ]].sum()
health_df.head()
```

```python
df.head(5)
```

```python
health_df['loggedDays']= df.groupby('Id')['ActivityDate'].count().sort_values(ascending = False)
```

```python
health_df['AvgCalorie']= health_df['Calories']/health_df['loggedDays']
```

```python
health_df['AvgHealthindex']= health_df['HealthyDay']/health_df['loggedDays']
```

```python
 health_df.sort_values(by=['AvgCalorie','HealthyDay'],ascending=False).head(3)
```

```python
health_df.describe()         # some statistics have done
```

```python
# classifyting a health status
def Health_Status(a,b):
    if ((a > 1800.0) or (b >0.8)):
        return 'Healthy'
    else:
        return 'UnHelthy'
    
```

```python
health_df['Health_Status']= health_df.apply(lambda health_df:Health_Status(health_df['AvgCalorie'],health_df['AvgHealthindex']), axis=1)
```
